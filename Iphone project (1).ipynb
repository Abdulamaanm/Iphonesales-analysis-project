```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
iphone = pd.read_csv("apple.csv")
y= iphone["Discount Percentage"].sort_values(ascending=False)[0:15]
# Create the pie chart with adjusted startangle
plt.pie(y
    , 
    labels=iphone["Product Name"][0:15], 
    autopct='%1.1f%%', 
    startangle=140  # Adjust this angle as needed
)

plt.title("Discount Percentage by Product")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
iphone = pd.read_csv("apple.csv")

# Sort the DataFrame by MRP in descending order and select the top 5 rows
top5_products = iphone.sort_values(by="Mrp", ascending=False).head(5)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top5_products["Product Name"], top5_products["Mrp"], color='skyblue')

# Add labels and title
plt.xlabel("Product Name")
plt.ylabel("MRP")
plt.title("Top 5 Products with the Highest MRP")
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability

# Show the plot
plt.tight_layout()  # Adjust layout to make room for label rotation
plt.show()

```
